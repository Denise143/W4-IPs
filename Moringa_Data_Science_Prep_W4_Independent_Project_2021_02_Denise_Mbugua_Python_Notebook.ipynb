```
# The csv file has to first be uploaded to our colab files
# using the function read_csv the dataset will be previewed
# I will only preview the last 10 rows

autolib = pd.read_csv('Autolib_dataset (2).csv')
autolib
```

# Dropping columns

```
# Some columns are irrelevant to the analysis 
# we will drop them
autolib.drop(['Address', 'Public name', 'Scheduled at', 'Displayed comment', 'Status', 'Subscription status', 'Geo point'], axis=1, inplace=True)
autolib
```

```
autolib.drop(['Station type'], axis=1, inplace=True)
autolib
```

# DATA CLEANING

## 1. Validity check

```
# using the strip() function of str to remove extra spaces on the ID column

autolib['ID'] = autolib['ID'].str.strip()
autolib
```

## Accuracy check

```
# Making sure minute is not greater then 60

autolib = autolib[(autolib['minute'] < 60) & (autolib['hour'] < 24) & (autolib['month'] < 13) & (autolib['day'] < 31)]

autolib.head()
```

```
# equating cars to utilib, utilib 1.4 and bluecars

autolib['Cars'] = autolib['Bluecar counter']
autolib.drop(['Cars'], axis=1, inplace=True)
autolib
```

## Completeness an Consistency checks

```
# To check for null values in the columns, isnull() function is used
# We then drop null values if they are there

autolib.isnull().sum()
```

```
# Removing rows with "center" under the column kind since they do not have resources

autolib = autolib[autolib.Kind != 'CENTER' ]
autolib
```

```
# Removing rows with rental status as future since they do not have resources
# dropping the rental status column
autolib = autolib[autolib["Rental status"] != "future"]
autolib.drop(['Rental status'], axis=1, inplace=True)
autolib
```

```
# Checking for duplicated rows
# Dropping duplicates if any

autolib.duplicated().sum()
autolib = autolib.drop_duplicates()
autolib
```

## Uniformity checks

```
# Changing all the column names to be uniform


autolib.rename({'year': 'Year', 'month': 'Month', 'day': 'Day', 'hour': 'Hour', 'minute': 'Minute', 'Charge Slots': 'Charge slots', 'Charging Status': 'Charging status'}, axis=1, inplace=True)
autolib.columns
```

```
# Replacing the hyphen with an underscore 
autolib['ID'] = autolib['ID'].str.lower().str.replace('-', '_')
autolib.head()
```

## Outliers


```
# Finding values which are far away from the mean
# There are none
autolib.boxplot(column =['Bluecar counter', 'Utilib counter', 'Utilib 1.4 counter'], grid = False)
```

## Exporting the clean dataframe

```
# converting our clean dataset to a csv file 
autolib.to_csv('Electric_cars.csv')

```

# DATA ANALYSIS

```
# joining successive time measures to form a new column

autolib_df = pd.read_csv('Electric_cars.csv')
autolib_df['Date'] = pd.to_datetime(autolib_df[['Year', 'Month', 'Day', 'Hour', 'Minute']])

autolib_df
```

## 1.Most Popular(Bluecar):




### Hour for picking

```
# Checking for the most popular hour that cars are picked
# we groupby the hours and sort the values in descending order
# to get the most popular hour, we select only the first row

A1 = autolib_df.groupby(['Hour', 'Bluecar counter'])['Hour'].count().sort_values(ascending=False)
A1.head(1)
```

### Station

```
# The most popular station overall

A2 = autolib_df.groupby(['ID', 'Bluecar counter', 'Postal code'])['ID'].count().sort_values(ascending=False)
print(A2.head(1))
```

```
# The most popular station at the most popular picking hour
A3 = autolib_df[(autolib_df['Hour'] == 14)]
A3.groupby(['ID', 'Bluecar counter'])['ID'].count().sort_values(ascending=False).head(1)
```

### Postal Code

```
# The most popular postal code and if it corresponds to the most popular station
# The most popular postal code does not correspond to the most popular station

A4 = autolib_df.groupby(['Postal code', 'Bluecar counter'])['Postal code'].count().sort_values(ascending=False)
A4
```

## 2.Most popular(Utilib and Utilib 1.4):

### Pick up hour

```
# Most popular pick up hour for utilib cars
A6 = autolib_df.groupby(['Hour', 'Utilib counter'])['Hour'].count().sort_values(ascending=False)
A6.head(1)
```

```
# Most popular pick up time for utilib1.4
A7 = autolib_df.groupby(['Hour', 'Utilib 1.4 counter'])['Hour'].count().sort_values(ascending=False)
A7.head(1)
```

### Station

```
# grouping to get the most used pick up station for utilib cars
# the values change for utilib and utilib 1.4 cars

A8 = autolib_df.groupby(['ID', 'Utilib counter', 'Postal code'])['ID'].count().sort_values(ascending=False)
print(A8.head(1))
```

```
# Getting the most used station for utilib 1.4

A9 = autolib_df.groupby(['ID', 'Utilib 1.4 counter', 'Postal code'])['ID'].count().sort_values(ascending=False)
print(A9.head(1))
```

### Postal code

```
# The most popular postal code and whether it is the code for the most popular pick up station
# It does not correspond
A10 = autolib_df.groupby(['Postal code', 'Utilib counter'])['Postal code'].count().sort_values(ascending=False)
A10
```

```
# Utilib 1.4

A11 = autolib_df.groupby(['Postal code', 'Utilib 1.4 counter'])['Postal code'].count().sort_values(ascending=False)
A11
```
